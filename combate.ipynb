```python
#Monitoramento dos dados de conteúdos e atividades
import pandas as pd
import numpy as np

# 1. Simulação de dados
np.random.seed(42)

# Criar um DataFrame simulado com dados de conteúdos e atividades
data = {
    'id_conteudo': range(1, 101),
    'tipo_conteudo': np.random.choice(['Filme', 'Música', 'Série', 'Software'], 100),
    'pirataria_detectada': np.random.choice([0, 1], 100, p=[0.8, 0.2]),  # 20% chance de pirataria
    'views': np.random.randint(1, 1000, 100),
    'downloads': np.random.randint(1, 500, 100),
}

df = pd.DataFrame(data)

# 2. Análise de dados
def analisar_pirataria(df):
    # Calcular a porcentagem de conteúdos piratas
    total_pirataria = df['pirataria_detectada'].sum()
    total_conteudos = df.shape[0]
    porcentagem_pirataria = (total_pirataria / total_conteudos) * 100

    print(f'Total de Conteúdos: {total_conteudos}')
    print(f'Total de Conteúdos Piratas: {total_pirataria}')
    print(f'Porcentagem de Pirataria: {porcentagem_pirataria:.2f}%')

    # Identificar os conteúdos mais visualizados e baixados
    conteudos_suspeitos = df[df['pirataria_detectada'] == 1]
    conteudos_suspeitos = conteudos_suspeitos.sort_values(by='views', ascending=False).head(5)

    print("\nConteúdos Suspeitos (mais visualizados):")
    print(conteudos_suspeitos[['id_conteudo', 'tipo_conteudo', 'views']])

    return conteudos_suspeitos

# 3. Sistema de alerta
def gerar_alertas(conteudos_suspeitos):
    for _, row in conteudos_suspeitos.iterrows():
        print(f'ALERTA: Conteúdo ID {row["id_conteudo"]} ({row["tipo_conteudo"]}) está sob suspeita de pirataria com {row["views"]} visualizações!')

# Executar a análise
conteudos_suspeitos = analisar_pirataria(df)

# Gerar alertas para os conteúdos suspeitos
gerar_alertas(conteudos_suspeitos)
```

```python
# 1. Adicionando dados de data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Exemplo de criação de um DataFrame com 657 entradas para 2023
data_2023 = {
    'tipo_conteudo': ['filme', 'serie', 'filme', 'serie'] * 164 + ['filme'],  # 657 entradas
    'pirataria_detectada': [5, 3, 7, 2] * 164 + [6]  # Total de 657
}

# Dados fictícios para 2022 e 2024
data_2022 = {
    'tipo_conteudo': ['filme', 'serie'] * 164 + ['filme'],  # 328 entradas
    'pirataria_detectada': [4, 5] * 164 + [7]  # Total de 328
}

data_2024 = {
    'tipo_conteudo': ['filme', 'serie'] * 164 + ['filme'],  # 328 entradas
    'pirataria_detectada': [6, 4] * 164 + [5]  # Total de 328
}

# Criando DataFrames
df_2023 = pd.DataFrame(data_2023)
df_2022 = pd.DataFrame(data_2022)
df_2024 = pd.DataFrame(data_2024)

# Adicionando a coluna de ano
df_2023['ano'] = 2023
df_2022['ano'] = 2022
df_2024['ano'] = 2024

# Concatenando os DataFrames
df = pd.concat([df_2022, df_2023, df_2024], ignore_index=True)

# Resumindo dados por ano e tipo de conteúdo
df_resumo = df.groupby(['ano', 'tipo_conteudo']).agg({'pirataria_detectada': 'sum'}).reset_index()

# Plotando o gráfico de barras agrupadas
plt.figure(figsize=(12, 6))
sns.barplot(data=df_resumo, x='ano', y='pirataria_detectada', hue='tipo_conteudo', palette='Set2')

# Título e rótulos
plt.title('Comparação de Pirataria por Tipo de Conteúdo (2022, 2023 e 2024)', fontsize=16)
plt.ylabel('Total de Pirataria Detectada', fontsize=14)
plt.xlabel('Ano', fontsize=14)
plt.xticks(rotation=0, fontsize=12)  # Rotação para melhor legibilidade

# Adicionando linhas de grade
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Melhorando a legenda
plt.legend(title='Tipo de Conteúdo', title_fontsize='13', fontsize='12')

# Definindo os limites do eixo Y para uma melhor visualização
plt.ylim(0, df_resumo['pirataria_detectada'].max() + 5)

# Salvando o gráfico
plt.savefig('comparacao_pirataria_anual.png', bbox_inches='tight')

plt.show()





```
